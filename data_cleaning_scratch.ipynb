```python
import re


# Define a function to extract the number from the URL
def extract_number(url):
    match = re.search(r'/(\d+)[.-]', url)  # Regex pattern to capture the number
    if match:
        return int(match.group(1))  # Return the captured number as integer
    else:
        return None

gr_100k_subset2 = gr_100k_subset.copy()

# Apply the function to extract the number from each URL in the 'link' column
gr_100k_subset2['link'] = gr_100k_subset2['link'].apply(extract_number)

# Rename the 'link' column to 'extracted_number'
gr_100k_subset2.rename(columns={'link': 'book_id'}, inplace=True)

gr_100k_subset2.head()
```

```python
# Function to concatenate sentences and remove numbers and quotes
def concatenate_sentences(sentences):
    concatenated_text = ''
    for sentence_info in sentences:
        # Remove the number at the beginning of each sentence
        sentence = sentence_info[1]
        if sentence_info[0] == 1:
            sentence = sentence[1:]  # Remove the number at the beginning
        # Remove quotes around the sentence
        if sentence.startswith('"') and sentence.endswith('"'):
            sentence = sentence[1:-1]
        # Append the sentence to the concatenated text
        concatenated_text += sentence + ' '
    return concatenated_text.strip()  # Remove trailing space

# Apply the function to concatenate sentences in the 'review_sentences' column
gr_reviews['review_sentences'] = gr_reviews['review_sentences'].apply(concatenate_sentences)
gr_reviews['review_sentences'][2]
```

```python
#Trying ijson
import ijson
import gzip

# Define the columns you want to keep
cols = ['isbn', 'language_code', 'average_rating', 'similar_books', 'description', 'link', 'authors', 'num_pages', 'book_id', 'ratings_count', 'title']

# Define the file name
file_name = 'data/goodreads_books.json.gz'

# Function to parse JSON records in chunks
def parse_json_chunk(file_name, cols, chunk_size=10000):
    data = []
    with gzip.open(file_name, 'rb') as f:
        parser = ijson.items(f, 'item')
        for idx, record in enumerate(parser):
            if idx % chunk_size == 0 and data:
                yield pd.DataFrame(data, columns=cols)
                data = []
            data.append([record.get(col) for col in cols])
    if data:
        yield pd.DataFrame(data, columns=cols)

# Read JSON file in chunks and concatenate into a single DataFrame
gr_books = pd.concat(parse_json_chunk(file_name, cols), ignore_index=True)
```
